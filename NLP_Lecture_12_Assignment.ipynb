```python
1. 请写一下TF-IDF的计算公式
```

$$ TF_{w_i, D_j} = \frac{count(w_i)}{|D_j|} 其中：count(w_i)为词w_i在文本D_j中出现的次数，|D_j|为D_j文本中所有词的数量.$$
$$IDF_{w_i} = log\frac{N}{1+\sum_{i=1}^{N}(I(w_i, D_i))} 其中：N为所有文本数量，\sum_{i=1}^{N}(I(w_i, D_i))表示出现词w_i的文档总数量$$ 
$$TF_IDF = TF_{w_i, D_j}*IDF_{w_i}$$

```python
##Text_rank实现1st:文本处理
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, TfidfTransformer
import jieba.posseg as pseg
import re

##生成词汇列表
data_source = "./sqlResult_1558435_1.csv"
data = pd.read_csv(data_source,encoding='gb18030')
data = data['Contents'].tolist()
#punctuations = ['?', '!', ';', '？', '！', '。', '；', '……', '…', '\n']
stopwords = [line.strip() for line in open('./stopwords.txt', 'r', encoding='utf-8').readlines()] ##+ list(punctuations)
speech_tags = ['an', 'i', 'j', 'l', 'n', 'nr', 'nrfg', 'ns', 'nt', 'nz', 't', 'v', 'vd', 'vn', 'eng']
news_content = [re.findall(r'[\d|\w]+',string) for string in data]
news_content = [''.join(n) for n in news_content]
news_words_lists = []
for text in news_content:
    seg_result = pseg.cut(text)
    text_word_list = [s.word for s in seg_result if s.flag in speech_tags and s.word not in stopwords]
    text_word_list = ' '.join(text_word_list)
    news_words_lists.append(text_word_list)
print(news_words_lists)
```

```python
##使用Tfidf生成text向量
vectorizer = TfidfVectorizer(max_features=10000)
sub_samples = news_words_lists
X = vectorizer.fit_transform(sub_samples)
#print(len(X.toarray()))
#print(vectorizer)
#print(vectorizer.get_feature_names())
vocabularies = vectorizer.vocabulary_
print(vocabularies)
#print(X.toarray()[0])
np.where(X[77].toarray())
```
